Plots for the report

```python
from matplotlib import pyplot as plt
import pandas as pd

list_df = []

df1910 = pd.read_csv("./csv/freeway_1910_returns.csv")
df42   = pd.read_csv("./csv/freeway_42_returns.csv")
df3047 = pd.read_csv("./csv/freeway_3047_returns.csv")


list_df.extend((df1910,df42,df3047))
list_colors = ['#e52592', 'green', '#f9ab00']
#list_colors = ['green', '#f9ab00']

list_labels = ["Extrinsic", "Intrinsic"]
```

```python
ax = plt.Axes

for n, df in enumerate(list_df):
    color = list_colors[n]
    smoothed_values = df['Value'].rolling(20).mean()
    plt.plot(df["Step"], df["Value"], alpha=0.1, color=color)
    plt.plot(df["Step"], smoothed_values, color=color)

plt.legend()    
plt.show()
```
